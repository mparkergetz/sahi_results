```python
import pandas as pd
```

```python
sahi_detections = pd.read_csv('csv/072424_detections_combined.csv')

sahi_detections.sort_values('filename').reset_index(drop=True)
```

```python
sahi_detections[['pi', 'date', 'timestamp']] = sahi_detections['filename'].str.extract(r'pi(\d+)_(\d{8})_(\d{6})')
sahi_detections['pi'] = sahi_detections['pi'].astype(int)
sahi_detections['datetime'] = pd.to_datetime(sahi_detections['date'] + sahi_detections['timestamp'], format='%Y%m%d%H%M%S')
sahi_detections = sahi_detections.drop(columns=['label', 'date', 'timestamp'])
sahi_detections
```

```python
detections_2024 = pd.read_csv('csv/2024_detections.csv', index_col=0)
```

```python
detections_2024[['pi', 'datetime']] = detections_2024['filename'].str.extract(r'pi(\d+)_([0-9]{8}_[0-9]{6})')
detections_2024['pi'] = detections_2024['pi'].astype(int)

detections_2024['datetime'] = pd.to_datetime(detections_2024['datetime'], format='%Y%m%d_%H%M%S')
detections_2024
```

```python
for df in [detections_2024, sahi_detections]:
    df['date'] = df['datetime'].dt.date


detections_2024['source'] = 'detections_2024'
sahi_detections['source'] = 'sahi_detections'
combined = pd.concat([detections_2024, sahi_detections], ignore_index=True)
grouped = combined.groupby(['pi', 'date', 'datetime', 'source']).size().unstack(fill_value=0)

def categorize(row):
    in_2024 = row.get('detections_2024', 0)
    in_sahi = row.get('sahi_detections', 0)
    if in_2024 and in_sahi:
        return 'shared'
    elif in_2024:
        return 'only_in_detections_2024'
    else:
        return 'only_in_sahi_detections'

grouped['category'] = grouped.apply(categorize, axis=1)
summary = grouped.reset_index().groupby(['pi', 'date', 'category']).size().unstack(fill_value=0).reset_index()
summary = summary.rename(columns={
    'shared': 'shared',
    'only_in_detections_2024': 'only_in_detections_2024',
    'only_in_sahi_detections': 'only_in_sahi_detections'
})

for col in ['shared', 'only_in_detections_2024', 'only_in_sahi_detections']:
    if col not in summary.columns:
        summary[col] = 0

summary
```

```python
summary[(summary['date'] >= pd.to_datetime('2024-07-20').date()) & (summary['date'] <= pd.to_datetime('2024-07-23').date())].to_csv('detections_0724_2.csv')
```

### SPLIT ALL_IMGS (FLATTENED SOURCE DIR) INTO OUTPUT SUBDIRS:
- BOTH
- 2024_DETECTIONS ONLY
- SAHI_ONLY
 - BIN BASED ON CONF SCORE


```python
import os
from pathlib import Path
import shutil
from multiprocessing import Pool, cpu_count

source_dir = "/home/misha/data/bee_cam/img_reviewing/all_imgs"
output_dir = "/home/misha/data/bee_cam/img_reviewing/output"
start_date = pd.to_datetime("2024-07-20")
end_date = pd.to_datetime("2024-07-23")

base_dirs = [
    "only_sahi/score_gt_0.75",
    "only_sahi/score_gt_0.5",
    "only_sahi/score_gt_0.25",
    "only_sahi/score_le_0.25",
    "only_2024",
    "both"
]

for d in base_dirs:
    os.makedirs(os.path.join(output_dir, d), exist_ok=True)

sahi_filtered = sahi_detections[
    (sahi_detections['datetime'] >= start_date) & 
    (sahi_detections['datetime'] <= end_date)
].copy()

detections_filtered = detections_2024[
    (detections_2024['datetime'] >= start_date) & 
    (detections_2024['datetime'] <= end_date)
].copy()

sahi_filtered['filename_ext'] = sahi_filtered['filename'] + '.jpg'
detections_filtered['filename_ext'] = detections_filtered['filename']

sahi_files = set(sahi_filtered['filename_ext'])
det_files = set(detections_filtered['filename_ext'])

only_sahi = sahi_files - det_files
only_det = det_files - sahi_files
both = sahi_files & det_files

def get_score_bin(score):
    if score > 0.75:
        return 'score_gt_0.75'
    elif score > 0.5:
        return 'score_gt_0.5'
    elif score > 0.25:
        return 'score_gt_0.25'
    else:
        return 'score_le_0.25'

sahi_only_df = sahi_filtered[sahi_filtered['filename_ext'].isin(only_sahi)].copy()
sahi_only_df['score_bin'] = sahi_only_df['score'].apply(get_score_bin)
sahi_only_df['subfolder'] = 'only_sahi/' + sahi_only_df['score_bin']

copy_tasks = []

for _, row in sahi_only_df.iterrows():
    copy_tasks.append((
        Path(source_dir) / row['filename_ext'],
        Path(output_dir) / row['subfolder'] / row['filename_ext']
    ))

for filename in only_det:
    copy_tasks.append((
        Path(source_dir) / filename,
        Path(output_dir) / 'only_2024' / filename
    ))

for filename in both:
    copy_tasks.append((
        Path(source_dir) / filename,
        Path(output_dir) / 'both' / filename
    ))

def copy_pair(src_dest):
    src, dest = src_dest
    if src.exists():
        shutil.copy2(src, dest)

with Pool(cpu_count()) as pool:
    pool.map(copy_pair, copy_tasks)
```

### SELECT OUT LARGE BBOXES

```python
import ast

AREA_THRESH = 10_000
AREA_SUBDIR = "area_gt_10000"

def parse_xywh(b):
    if isinstance(b, (list, tuple)):
        return b
    return list(ast.literal_eval(str(b)))

sahi_only_df["xywh"] = sahi_only_df["bbox"].apply(parse_xywh)
sahi_only_df["area_px"] = sahi_only_df["xywh"].apply(lambda v: float(v[2]) * float(v[3]))

files_over = (
    sahi_only_df.loc[sahi_only_df["area_px"] > AREA_THRESH, ["filename_ext", "score_bin"]]
    .drop_duplicates()
)

extra_copy_tasks = []
for _, row in files_over.iterrows():
    score_bin = row["score_bin"]
    fname = row["filename_ext"]

    # source = where we put them earlier:
    src = Path(output_dir) / "only_sahi" / score_bin / fname
    # destination = subdir in same score folder:
    dst_dir = Path(output_dir) / "only_sahi" / score_bin / AREA_SUBDIR
    dst_dir.mkdir(parents=True, exist_ok=True)
    dst = dst_dir / fname

    extra_copy_tasks.append((src, dst))

# Parallel copy
with Pool(cpu_count()) as pool:
    pool.map(copy_pair, extra_copy_tasks)

print(f"Copied {len(extra_copy_tasks)} images into '{AREA_SUBDIR}' subfolders (threshold={AREA_THRESH} px^2).")
```
