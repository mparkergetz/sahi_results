```python
import pandas as pd
```

```python
sahi_detections = pd.read_csv('csv/detections.csv')

sahi_detections.sort_values('filename').reset_index(drop=True)
```

```python
sahi_detections[['pi', 'date', 'timestamp']] = sahi_detections['filename'].str.extract(r'pi(\d+)_(\d{8})_(\d{6})')
sahi_detections['pi'] = sahi_detections['pi'].astype(int)
sahi_detections['datetime'] = pd.to_datetime(sahi_detections['date'] + sahi_detections['timestamp'], format='%Y%m%d%H%M%S')
sahi_detections = sahi_detections.drop(columns=['label', 'date', 'timestamp'])
sahi_detections
```

```python
detections_2024 = pd.read_csv('csv/2024_detections.csv', index_col=00
```

```python
detections_2024[['pi', 'datetime']] = detections_2024['filename'].str.extract(r'pi(\d+)_([0-9]{8}_[0-9]{6})')
detections_2024['pi'] = detections_2024['pi'].astype(int)

detections_2024['datetime'] = pd.to_datetime(detections_2024['datetime'], format='%Y%m%d_%H%M%S')
detections_2024
```

```python
for df in [detections_2024, sahi_detections]:
    df['date'] = df['datetime'].dt.date


detections_2024['source'] = 'detections_2024'
sahi_detections['source'] = 'sahi_detections'
combined = pd.concat([detections_2024, sahi_detections], ignore_index=True)
grouped = combined.groupby(['pi', 'date', 'datetime', 'source']).size().unstack(fill_value=0)

def categorize(row):
    in_2024 = row.get('detections_2024', 0)
    in_sahi = row.get('sahi_detections', 0)
    if in_2024 and in_sahi:
        return 'shared'
    elif in_2024:
        return 'only_in_detections_2024'
    else:
        return 'only_in_sahi_detections'

grouped['category'] = grouped.apply(categorize, axis=1)
summary = grouped.reset_index().groupby(['pi', 'date', 'category']).size().unstack(fill_value=0).reset_index()
summary = summary.rename(columns={
    'shared': 'shared',
    'only_in_detections_2024': 'only_in_detections_2024',
    'only_in_sahi_detections': 'only_in_sahi_detections'
})

for col in ['shared', 'only_in_detections_2024', 'only_in_sahi_detections']:
    if col not in summary.columns:
        summary[col] = 0

summary
```
